### Create function to find area of circle given radius or diameter, based on user choice:


```python
import math

def area():
    
    choice = ''
    
    while choice != 'radius' and choice != 'diameter':
        choice = input('Enter radius or diameter: ')
        if choice != 'radius' or choice != 'diameter':
            print('You must enter radius or diameter')
            
    if choice == 'radius':
        radius = input('Radius length in inches: ')
        area = math.pi * float(radius) * float(radius)
        print('The area of the circle is {0}'.format(area))
    
    if choice == 'diameter':
        diameter = input('Diameter length in inches: ')
        area = math.pi * (float(diameter/2) * float(diameter/2))
        print('The area of the circle is {0} inches'.format(area))

```

```python
area()
```

```python
# Triangle Class Exercise to check if angles are 180 or not

class Triangle():
    def __init__(self, angle1, angle2, angle3):
        
        self.angle1 = angle1
        self.angle2 = angle2
        self.angle3 = angle3
        
    number_of_sides = 3
        
    def check_angles(self):
            
        if self.angle1 + self.angle2 + self.angle3 == 180:
                
            return True
        else:
                
            return False     
            

# Inputs Test
my_triangle = Triangle(30,50,100)

print(my_triangle.number_of_sides)

print(my_triangle.check_angles())
```

## Alternatives to Nested ElseIf statements

```python
# dict used instead of if else setatements

price_list = {
'fish': 8,
'beef': 7,
'broccoli': 3,
}
```

```python
# if else function

def find_price(item):
    if item in price_list:
        return f'The price for {item} is {price_list[item]}'
    else:
        return f'The price for {item} is not available'

print(find_price('fish'))

print(find_price('cauliflower'))
```

```python
# function replace if else

def find_price(item):
    return f'The price for {item} is {price_list.get(item, "not available")}'

print(find_price('fish'))

print(find_price('cauliflower'))
```

## Generators and Context Managers

### Generators: By using generators paired with 'yield' it is possible to temporarily use memory instead of loading all of it into memory, used for both functions and lists

### Context Managers: Good use to free up computing resources instead of storing connectinos in memory, or storing large lists of values in memory.

### Generators

```python
# Iterator on a list - set up for Generators

a_list = list(range(0,4))

print(a_list)

list_iter = iter(a_list)

next(list_iter)
```

```python
next(list_iter)
```

```python
# Generator Function - they yield a value rather than returning a value, state is preserved between yields
# Generators are really iterators that do the 'bookkeeping' for you, used to iterate over integers since they are not iterable by themselves

def y_range(start, stop, steps=1):
    i = start
    
    while i < stop:
        yield i
        i += steps
```

```python
y = y_range(3,5)
```

```python
y
```

```python
next(y)
```

## Syntax - Decorators

```python
# simple add function to show use of decorators, decorators are functions that take other 
# functions as inputs

def add(a,b):
    print(f'Function "add" called with args: {a}, {b}')
    result = a + b
    
    print(f'Result: {result}')
    return result

```

```python
add(4,5)
```
